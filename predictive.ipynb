```python
import pandas as pd
```

```python
df = pd.read_csv('nba_games.csv', index_col=0)
df
```

```python
df = df.sort_values("date")
```

```python
df = df.reset_index(drop=True)
```

```python
df
```

```python
del df["mp.1"]
del df["mp_opp.1"]
del df["index_opp"]
```

```python
df
```

```python
# target indicates if the team lost their next game

def add_target(team):
    team ["target"] = team["won"].shift(-1)
    return team

df = df.groupby("team", group_keys=False).apply(add_target)
```

```python
df[df["team"] == "BOS"]
```

```python
df["target"][pd.isnull(df["target"])]
```

```python
def true_false_conversion(x):
    if x == True:
        return 1
    elif x == False:
        return 0
    else:
        return 2
df["target"] = df["target"].apply(true_false_conversion)
```
